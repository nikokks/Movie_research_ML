```python
from transformers import AutoTokenizer, AutoModel
import torch
os.environ['TORCH_HOME'] = 'models/' #setting the environment variable
tokenizer = AutoTokenizer.from_pretrained("google/bigbird-roberta-base")
model = AutoModel.from_pretrained("google/bigbird-roberta-base").to('cuda')
```

```python
import pandas as pd
df= pd.read_csv('wiki_movie_plots_deduped.csv')
```

```python
def encoding(liste):
    encoded_input = tokenizer(liste, padding=True, truncation=True, max_length=1024, return_tensors='pt').to('cuda')
    # compute token embeddings
    with torch.no_grad():
        model_output = model(**encoded_input).pooler_output.cpu()
    return  np.array(model_output)
```

```python
liste_output = []
liste_text = df.Plot.tolist()
nb_batches=32
for i in range(len(liste_text)//nb_batches+1):
    #liste = liste_text[:1]
    liste = liste_text[i*nb_batches:(i+1)*nb_batches]
    liste_output = liste_output + encoding(liste).cpu().tolist()
    print(i,len(liste_text)//nb_batches+1)
```
